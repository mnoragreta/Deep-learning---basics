## Hyperparameter tuning

### Tuning 1 - Hyperband

No of neuron in layers and learning rate to be tuned

```python
def build_model(hp):
    dl_model5 = Sequential()
    dl_model5.add(Dense(units=hp.Int('units_1', min_value=16, max_value=128, step=16), activation='relu', input_shape=(X_train.shape[1],)))
    dl_model5.add(Dense(units=hp.Int('units_2', min_value=16, max_value=128, step=16), activation='relu'))
    dl_model5.add(Dense(1, activation='linear'))
    
    dl_model5.compile(
        optimizer=Adam(learning_rate=hp.Float('learning_rate', min_value=1e-4, max_value=1e-2, sampling='LOG')),
        loss='mean_squared_error',
        metrics=[tf.keras.metrics.RootMeanSquaredError()]
    )
    
    return dl_model5

tuner = kt.Hyperband(
    build_model,
    objective='val_root_mean_squared_error',
    max_epochs=10,
    factor=3,
    directory='C:/Users/Lily/Documents/keras_tuner',
    project_name='helloworld'
)

tuner.search(X_train, y_train, epochs=10, validation_data=(X_test, y_test))

best_model = tuner.get_best_models(num_models=1)[0]
best_hyperparameters = tuner.get_best_hyperparameters(num_trials=1)[0]
```

```python
best_learning_rate = best_hyperparameters.get('learning_rate')
best_units_1 = best_hyperparameters.get('units_1')
best_units_2 = best_hyperparameters.get('units_2')

print(f"Best Learning Rate: {best_learning_rate}")
print(f"Best Number of Neurons in First Hidden Layer: {best_units_1}")
print(f"Best Number of Neurons in Second Hidden Layer: {best_units_2}")
```

### Tuning 2 - ParamGrid tuning for epochs and batch_size

Epochs and batch_size to be tuned

```python
param_dist = {
    'epochs': [50, 70, 100],
    'batch_size': [16, 32, 64]  
}
```

```python
best_rmse = float('inf')
best_params = None

for params in ParameterGrid(param_dist):
    dl_model6 = Sequential()
    dl_model6.add(InputLayer(input_shape=(15,)))  

    dl_model6.add(Dense(96, activation='relu'))
    dl_model6.add(Dense(32, activation='relu'))
    dl_model6.add(Dense(1, activation='linear'))

    dl_model6.compile(optimizer=Adam(learning_rate=0.001), loss='mse', metrics=[tf.keras.metrics.RootMeanSquaredError()])

    early_stopping = EarlyStopping(monitor='val_loss', patience=5, min_delta=0.001, restore_best_weights=True)

    history6 = dl_model6.fit(X_train, y_train,
                    validation_data=(X_test, y_test),
                    epochs=params['epochs'],
                    batch_size=params['batch_size'],
                    callbacks=[early_stopping],
                    verbose=1)
    
    val_rmse = min(history6.history['val_root_mean_squared_error'])


    if val_rmse < best_rmse:
        best_rmse = val_rmse
        best_params = params

print(f'Best RMSE: {best_rmse}')
print(f'Best Parameters: {best_params}')
```
